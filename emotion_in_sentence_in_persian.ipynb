```python
from keras.layers import Dense , Embedding , SimpleRNN
from keras.models import Sequential
```

```python
from keras.datasets import imdb
import numpy as np
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
numwords  = 15000
(x_train , y_train), (x_test , y_test) = imdb.load_data(num_words= numwords , maxlen=300)
```

```python
word_index = imdb.get_word_index()
```

```python
def number_to_word(comment):
    sentence = ''
    for i in comment:
        for (word,number) in word_index.items():
            if number == i:
               sentence = sentence + ' ' + word
               break
    print(sentence) 
```

```python
number_to_word(x_train[0])
```

```python
x_train = pad_sequences(x_train , maxlen=300)
x_test = pad_sequences(x_test , maxlen=300)
```

```python
model = Sequential()

model.add(Embedding(numwords , 32 , input_length=300  ))

model.add(SimpleRNN(32 , input_shape = (numwords , 300) ))

model.add(Dense(1))




model.compile(loss='binary_crossentropy' , optimizer='rmsprop' , metrics=['accuracy'])
```

```python
model.summary()
```

```python
model.fit(x_train , y_train , validation_data=(x_test , y_test) , epochs=10)
```

```python
model.save_weights('jomle.h5')
```
